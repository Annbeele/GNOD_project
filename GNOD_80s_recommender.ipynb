#### Get album ids

```python
def get_ablum_ids_from_artist_ids(artist_ids):
    
    album_ids=[]
    for i in artist_ids:
        albums_info = sp.artist_albums(i, album_type='album')
        albums_info_items = albums_info['items']
        # print(albums_info_items)   --->  [{'album_group': 'album', 'album_type': 'album', 'artists': [{'external_urls': {'spotify': 'https://open.spotify.com/artist/0lZoBs4Pzo7R89JM9lxwoT'}, 'href': 'https://api.spotify.com/v1/artists/0lZoBs4Pzo7R89JM9lxwoT', 'id': '0lZoBs4Pzo7R89JM9lxwoT', 'name': 'Duran Duran',
        while albums_info['next']:
            albums_info = sp.next(albums_info)
            albums_info_items.extend(albums_info['items'])
        for i in albums_info_items:
            album_ids.append(i['id'])
    
    return album_ids
```

#### Get all song names from album ids

```python
def get_song_names_from_album_ids(album_ids):
    
    song_names=[]
    for i in album_ids:
        albums_tracks= sp.album_tracks(i)
        albums_tracks_items = albums_tracks['items']
        # print(albums_info_items)   --->  [{'album_group': 'album', 'album_type': 'album', 'artists': [{'external_urls': {'spotify': 'https://open.spotify.com/artist/0lZoBs4Pzo7R89JM9lxwoT'}, 'href': 'https://api.spotify.com/v1/artists/0lZoBs4Pzo7R89JM9lxwoT', 'id': '0lZoBs4Pzo7R89JM9lxwoT', 'name': 'Duran Duran',
        while albums_tracks['next']:
            albums_tracks = sp.next(albums_tracks)
            albums_tracks_items.extend(albums_tracks['items'])
        for i in albums_tracks_items:
            song_names.append(i['name'])
    
    return song_names
```

- Changing parameters to look for the lowest inertia

```python
# Try to run Kmeans with all values of K, from 55 to 65 (we have tested before 25-45 , 35-45, 45-55 ...)
K = range(55, 65)

# For each model, store the inertia in a list
inertia = []

for k in K:
    kmeans=KMeans(n_clusters=k, random_state=40)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)
    
print(inertia)
```

```python
# Plot the results: Elbow method
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
# Silhouette method
from sklearn.metrics import silhouette_score

K = range(55, 65)

silhouettes = []

for k in K: 
    kmeans=KMeans(n_clusters=k, random_state=40)
    kmeans.fit(X_prep)
    silhouettes.append(silhouette_score(X_prep, kmeans.predict(X_prep)))
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(16,8))
plt.plot(K, silhouettes, 'bo-')
plt.xlabel('k (number of clusters)')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.ylabel('silhouette score')
```

- Comparing Elbow method and Silhouette and considering we got 20.000 songs, we decide to choose **k=63**

```python
kmeans = KMeans(n_clusters=63,
                init="random",
                n_init=10,  
                max_iter=200,
                tol=0,
                algorithm="full",
                random_state=40)
kmeans.fit(X_prep)
print(kmeans.inertia_)
```

### BUILDING THE RECOMMENDER

```python
def RECOMMENDER_80s():
    
    from sklearn.preprocessing import StandardScaler
    from sklearn.cluster import KMeans
    import random
    import pandas as pd
    import spotipy
    from spotipy.oauth2 import SpotifyClientCredentials
    
    print('Welcome to the 80s SONG RECOMMENDER! Tell me a song you like and we will do the rest...')
    
    song = input("What song do you like? ")
    artist = input("From which artist? ")
    
    user_result = sp.search(q="artist:" + artist + " track:" + song, type="track")
    
    if user_result['tracks']['items'] == []:
        print ('Oops, try another song or check for typos')
        
    else:
        song_id = user_result['tracks']['items'][0]['id']
        user_features = pd.DataFrame(sp.audio_features(tracks=song_id))
        user_af = user_features.drop(columns = ['key','mode','type','id','uri','track_href','analysis_url','duration_ms','time_signature'])

        X = pd.read_csv('X_af_ref_to_scale.csv')
        X = X.drop(['Unnamed: 0'], axis = 1)

        Scaler = StandardScaler()
        Scaler.fit(X)
        user_af_scaled = Scaler.transform(X)

        kmeans = KMeans(n_clusters=63,
                    init="random",
                    n_init=10,  
                    max_iter=200,
                    tol=0,
                    algorithm="full",
                    random_state=40)
        kmeans.fit(user_af_scaled)

        user_cluster = kmeans.predict(user_af_scaled)
        user_cluster = user_cluster[0]

        playlist_ref = pd.read_csv('clustered_80s_playlist.csv')
        playlist_ref.drop('Unnamed: 0', axis='columns', inplace=True)

        #subset of songs that equals user's cluster

        subset = playlist_ref[playlist_ref['cluster'] == user_cluster]
        subset = subset.reset_index(drop=True)

        song_titles = list(subset['title'])
        song_recom = random.choice(song_titles)

        index = list(subset['title']).index(song_recom)
        artist = list(subset.iloc[[index]]['artist'])[0]


        return print('You might also like:', song_recom,', by', artist )
```

```python
RECOMMENDER_80s()
```
