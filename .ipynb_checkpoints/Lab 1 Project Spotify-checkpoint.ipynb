```python

def billboard_scraper():
    from bs4 import BeautifulSoup
    import requests
    import pandas as pd
    import random
    # 2. find url and store it in avariable
    url = "https://www.billboard.com/charts/hot-100"
    # 3. download html with a get request
    response = requests.get(url)
    soup=BeautifulSoup(response.content, 'html.parser')
    titles=[]
    artists=[]
    for i in soup.find_all('span', {'class':'chart-element__information__song'}):
        titles.append(i.get_text())
    for i in soup.find_all('span', {'class':'chart-element__information__artist'}):
        artists.append(i.get_text())
    return pd.DataFrame({'song':titles, 'artist':artists})
```

```python
#para q me salgan todos los resultados hago:

def get_playlist_tracks(user_id, playlist_id):
    results = sp.user_playlist_tracks(user_id, playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

#### Splittingg the 34243 songs into groups of 50

```python
def track_chunks(song_ids):
    chunks = [song_ids[x:x+50] for x in range(0, len(song_ids), 50)]
    return chunks
```

```python
def track_chunks_100(song_ids):
    chunks = [song_ids[x:x+100] for x in range(0, len(song_ids), 100)]
    return chunks
```

### Function to apply audio features to 31.243 songs

```python
def audio_features_from_song_ids(song_ids):
    
    chunks = track_chunks_100(song_ids)
    
    audio_feat_list = []
    
    for i in chunks:
        audio_feat = sp.audio_features(tracks=i)
        audio_feat_list.append(audio_feat)
    
    flat_list = [i for sublist in audio_feat_list for i in sublist]
    
    return flat_list
```

### KMEANS

```python
from sklearn.cluster import KMeans

#define the model, fit the model to your data 

kmeans = KMeans(n_clusters=8, random_state=40)
kmeans.fit(X_prep)
```

```python
#look at the cluster centres 
kmeans.cluster_centers_
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_prep)
```

```python
# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
# Explore the cluster assignment by placing it in the original dataset
X_cluster=pd.DataFrame(X)
X_cluster['cluster']=clusters
X_cluster.head(30)
```

```python
#X_cluster.shape
```

```python
# checking total inertia of all the centroids:

kmeans.inertia_
```

```python
#changing parameters to look for the lowest inertia
```

```python
# Try to run Kmeans with all values of K, from 55 to 65 (we have tested before 25-45 , 35-45, 45-55 ...)
K = range(55, 65)

# For each model, store the inertia in a list
inertia = []

for k in K:
    kmeans=KMeans(n_clusters=k, random_state=40)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)
    
print(inertia)
```

```python
# Plot the results: Elbow method
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
# Silhouette method
from sklearn.metrics import silhouette_score

K = range(55, 65)

silhouettes = []

for k in K: 
    kmeans=KMeans(n_clusters=k, random_state=40)
    kmeans.fit(X_prep)
    silhouettes.append(silhouette_score(X_prep, kmeans.predict(X_prep)))
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(16,8))
plt.plot(K, silhouettes, 'bo-')
plt.xlabel('k (number of clusters)')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.ylabel('silhouette score')
```

- Comparing Elbow method and Silhouette and considering we got 20.000 songs, we decide to chooose **k=63**

```python
kmeans = KMeans(n_clusters=63,
                init="random",
                n_init=10,  
                max_iter=200,
                tol=0,
                algorithm="full",
                random_state=40)
kmeans.fit(X_prep)
print(kmeans.inertia_)
```
